```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv (r'Demographic Survey Data (for UrsaTech) - Form Responses 1.csv')
```

```python
df.dropna(inplace = True)
df.rename(columns = {'Are you a historically Underrepresented Ethnic Minority (URM)? [i.e. Black/African-American, Latinx/Chicanx, Filipinx, Native American/Indigenous, or Pacific Islander/Native Hawaiian]': 'URM'}, inplace = True)
df.drop(df[df.URM == 'Decline to say'].index, inplace = True)
```

```python
df.groupby(['URM','How Would You Rate the Climate of Clubs/Student Organizations at BERKELEY?']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Climate at Berkeley',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','How Would You Rate the Climate of Clubs/Student Organizations YOU ARE IN?']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Climate in Club',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','How Would You Rate the Inclusiveness of Clubs/Student Organizations at BERKELEY? ']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Inclusiveness at Berkeley',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','How Would You Rate the Inclusiveness of the Clubs/Student Organizations that YOU ARE IN? ']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Inclusiveness in Club',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','Do You Feel Welcome at Berkeley?']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Welcome at Berkeley',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','Do You Feel as if You Are Part of the Berkeley Community?']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Part of Berkeley Community',color = ['b', 'g','r','orange','c'])
```

```python
df.groupby(['URM','Do You Feel as if You Are Part of the Berkeley Engineering Community?']).size().unstack().plot(kind='bar', figsize = (10,5), title = 'Part of Engineering Community',color = ['b', 'g','r','orange','c'])
```
